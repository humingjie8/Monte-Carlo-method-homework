```python
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns


# 使用切片来获取前N项并计算它们的和
def sum_of_first_n_elements(arr, n):
    return sum(arr[:n])


# 设置随机数种子以确保可重复性
np.random.seed(42)

#设置总样本数据量
number_data = 100

# 生成符合正态分布N(0, 1)的随机数
mu, sigma = 0, 1  # 均值和标准差

standard_normal_samples = np.random.standard_normal(size=number_data)  # 生成10000个样本
mean_value = np.mean(standard_normal_samples)
min_value = np.min(standard_normal_samples)

print(mean_value, min_value)
t = [0] * number_data

for i in range(number_data):
    mid_1 = sum_of_first_n_elements((standard_normal_samples-mean_value)**2, i)
    t[i] = (math.sqrt(mid_1/(i+1)))/(mean_value-min_value)


# print(np.min(t))

# 计算分位数
N1 = np.percentile(t, 0.01)
N2 = np.percentile(t, 0.05)
N3 = np.percentile(t, 0.1)
print(N1, N2, N3)

# 绘制KDE图
sns.kdeplot(t, fill=True, label='KDE', bw_adjust=,color='r')
plt.title('Kernel Density Estimation (KDE) of Data')

# 设置横坐标和纵坐标的范围
plt.xlim(0, 1)  # 设置横坐标的范围为1到5
# plt.ylim(0, 12)  # 设置纵坐标的范围为0到12

plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 指数分布参数
lambda_param = 1

# 生成一组随机样本
samples = np.random.exponential(scale=1/lambda_param, size=1000)

# 绘制直方图
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g')

# 生成PDF曲线
x = np.linspace(0, 10, 100)
pdf = lambda_param * np.exp(-lambda_param * x)
plt.plot(x, pdf, 'r-', lw=2)

plt.title(f'Exponential Distribution (λ=1)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend(['PDF', 'Histogram'])
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 生成一组标准正态分布的随机数据
data = np.random.randn(1000)

# 计算标准正态分布的概率密度函数
mu, std = norm.fit(data)
pdf = norm.pdf(data, mu, std)

# 创建一个包含多个子图的图形
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# 绘制直方图
ax1.hist(data, bins=30, density=True, alpha=0.6, color='g')

# 绘制概率密度函数曲线
xmin, xmax = min(data), max(data)
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
ax1.plot(x, p, 'k', linewidth=2)

# 设置图的标题和标签
ax1.set_title("Histogram and PDF")
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Frequency")

# 绘制概率密度函数曲线
x = np.linspace(-5, 5, 100)
y = norm.pdf(x, 0, 1)
ax2.plot(x, y, 'r', linewidth=2)

# 设置图的标题和标签
ax2.set_title("Standard Normal Distribution PDF")
ax2.set_xlabel("X-axis")
ax2.set_ylabel("PDF")

# 显示图形
plt.show()
```
